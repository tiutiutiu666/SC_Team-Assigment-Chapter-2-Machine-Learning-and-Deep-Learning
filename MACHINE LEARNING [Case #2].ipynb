### Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn import cluster
from sklearn.metrics import silhouette_score
```

### Read Dataset

```python
data = pd.read_csv('https://raw.githubusercontent.com/Rietaros/kampus_merdeka/main/cluster_s1.csv')
data = data.drop('no', axis = 1)
data.head()
```

```python
%matplotlib inline
ax1 = data.plot.scatter(x='x', y='y', c='DarkBlue')
#plt.show()
```

### Data Segmentation (Clustering)

```python
# TODO: Lakukan pencarian jumlah cluster terbaik berdasarkan data diatas
# Menggunakan metode silhouette
silhouette_scores = []
K = range(2, 25)  # Mencari kluster terbaik dari 2 hingga 25

for k in K:
    kmeans = cluster.KMeans(n_clusters=k, random_state=0)
    kmeans.fit(data)
    labels = kmeans.labels_
    silhouette_scores.append(silhouette_score(data, labels))

# Menentukan jumlah cluster terbaik
best_n_clusters = K[np.argmax(silhouette_scores)]
print("Jumlah kluster terbaik: ", best_n_clusters)
```

```python
# TODO: Lakukan pemodelan dengan KMeans
kmeans = cluster.KMeans(n_clusters=best_n_clusters, random_state=0)
kmeans.fit(data)
label = kmeans.labels_

score = silhouette_score(data, label)
print("Silhouette score={:.5f}".format(score))
```

### Visualization: Cluster Result

```python
# TODO: Masukkan label hasil clustering ke dalam DataFrame
result = data.copy()
result['cluster'] = label
result.head()
```

```python
# TODO: Plot data hasil clustering dengan Seaborn
plt.figure(figsize=(10, 6))
sns.scatterplot(x='x', y='y', hue='cluster', data=result, palette='tab20')
plt.title(f'Clustering with {best_n_clusters} Clusters')
plt.show()
```

```python
g = sns.relplot(x='x', y='y', hue='cluster', data=result, palette='tab20', row='cluster', height=4, aspect=1.5)
g.set_titles('Cluster {row_name}')
plt.show()
```
