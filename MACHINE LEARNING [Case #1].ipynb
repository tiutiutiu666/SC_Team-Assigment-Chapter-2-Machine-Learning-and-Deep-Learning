### Import Library

```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split as tts
from sklearn.model_selection import GridSearchCV

from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

### Read Dataset

```python
df = pd.read_csv('https://raw.githubusercontent.com/Rietaros/kampus_merdeka/main/SC_HW1_bank_data.csv')
#df.columns
df.head(5)
```

```python
# TODO: Hilangkan kolom yang tidak relevan untuk pemodelan
#df = df[[...]].copy()
df = df.drop(columns=['RowNumber', 'CustomerId'])
df.head(10)
```

### Preprocessing

```python
# TODO: Lakukan One-Hot Encoder untuk data kategorikal, dengan fungsi pd.get_dummies
dummies1 = pd.get_dummies(df[['Gender']], drop_first=True, dtype=int)
dummies2 = pd.get_dummies(df[['Geography']],dtype=int)
df = pd.concat([df,dummies1, dummies2], axis=1).drop(columns=['Gender', 'Geography'], axis=1)
df
```

```python
# TODO: Pisahkan fitur (X) dengan target (Y), dimana Y merujuk ke kolom "Exited"
X = df.drop(['Exited'], axis=1)
y = df['Exited']
```

```python
scaler = MinMaxScaler() # TODO: Lakukan scaling atau normalisasi
X_transform = scaler.fit_transform(X)
```

```python
# Bagian ini digunakan jika Anda melakukan scaling atau normalisasi.
# Jika tidak, code ini bisa dilewati dan diganti dengan code yang ada di dalam comment.

X_transform = pd.DataFrame(X_transform, columns = X.columns)
X_transform.head()
# X_transform = X.copy()
```

### Train-Test Split

```python
X_train ,X_test, y_train, y_test = tts(X_transform, y, test_size = 0.25, random_state = 123)
```

### Model #1

#### Pertanyaan: Jelaskan secara singkat model #1 yang Anda gunakan.

**Logistic Regression**

Logistic Regression adalah model machine learning yang digunakan untuk data yang sifatnya kategori atau diskrit. Ketika kita ingin mengetahui/mengkategorikan kemungkinan suatu peristiwa terjadi atau tidak. Misalnya, memprediksi apakah seorang karyawan telah keluar dari perusahaan atau tidak.

```python
from sklearn.linear_model import LogisticRegression
# model1 = LogisticRegression()
# params = {"tol": [0.1,0.01,0.001], 'C':[0.5,1.0,1.5,2.0]}


# TODO: Pilih salah satu model Machine Learning (ML) dari Scikit-learn
model1 = LogisticRegression()
params =  {
    'C': [0.5,1.0,1.5,2.0],
    "tol": [0.1,0.01,0.001],
    'solver': ['sag']
}

# OPTIONAL: Lakukan hyperparameter tuning sesuai kebutuhan
grid = GridSearchCV(
    estimator= model1,
    param_grid= params,
    scoring = 'accuracy',
    n_jobs = 10,
    cv = 10 # 3-fold cross validation (artinya kita melakukan iterasi model sebanyak 3 kali)
)

grid.fit(X_train,y_train)
grid.best_params_
```

```python
y_pred = grid.predict(X_test) # TODO: Lakukan evaluasi untuk model #1


print(classification_report(y_test,y_pred))
print("")
print(confusion_matrix(y_test,y_pred))
print("")
print(accuracy_score(y_test, y_pred))
```

### Model #2

#### Pertanyaan: Jelaskan secara singkat model #2 yang Anda gunakan.

**Gradient Boosting Classifier**

Gradient Boosting Classifier adalah metode machine learning dengan cara menggunakan model lemah untuk membuat sebuah model ensembel yang lebih kuat. Karena metaode ini menggabungkan banyak model yang lebih sederhana (disebut weak learners), maka akan menghasilkan model yang jauh lebih kompleks dan akurat

```python
from sklearn.ensemble import GradientBoostingClassifier
# model1 = LogisticRegression()
# params = {"tol": [0.1,0.01,0.001], 'C':[0.5,1.0,1.5,2.0]}

# TODO: Pilih salah satu model Machine Learning (ML) dari Scikit-learn
model2 = GradientBoostingClassifier()
params2 = {
    'n_estimators':[1, 2, 5, 10, 20, 50, 100],
    'learning_rate':[0.2, 0.1, 0.05, 0.01],
    'max_depth':[3, 4, 5]
}

# OPTIONAL: Lakukan hyperparameter tuning sesuai kebutuhan
grid2 = GridSearchCV(
    estimator= model2,
    param_grid= params2,
    scoring = 'accuracy',
    n_jobs = 10,
    cv = 10
)

grid2.fit(X_train,y_train)
grid2.best_params_
```

```python
# TODO: Lakukan evaluasi untuk model #2
y_pred2 = grid2.predict(X_test)

print(classification_report(y_test, y_pred2))
print("")
print(confusion_matrix(y_test, y_pred2))
print("")
print(accuracy_score(y_test, y_pred2))
```

### Model #3

#### Pertanyaan: Jelaskan secara singkat model #3 yang Anda gunakan.

**Random Forest Classifier**

Random Forest Classifier adalah metode machine learning yang bekerja dengan memilih seluruh kemungkinan yang ada untuk menjadi sebuah root node, lalu menghitung seluruh kemungkinan dengan cara menggunakan model lemah untuk membuat sebuah model ensembel yang akan terbentuk berdasarkan root node yang dipilih.

```python
from sklearn.ensemble import RandomForestClassifier
model3 = RandomForestClassifier()
params3 = {
    'n_estimators':[50, 100, 150],
    'criterion':['gini', 'entropy'],
    'random_state':[8, 24],
    'max_depth':[None, 8, 12]
}

# OPTIONAL: Lakukan hyperparameter tuning sesuai kebutuhan
grid3 = GridSearchCV(
    estimator = model3,
    param_grid = params3,
    scoring = 'accuracy',
    n_jobs = 10,
    cv = 10
)

grid3.fit(X_train,y_train)
grid3.best_params_
```

```python
y_pred3 = grid3.predict(X_test)

print(classification_report(y_test,y_pred3))
print(confusion_matrix(y_test,y_pred3))
print(accuracy_score(y_test, y_pred3))
```

### Conclusion

#### Pertanyaan: Tarik kesimpulan model mana yang terbaik beserta alasannya?

Berdasarkan hasil evaluasi model, dapat terlihat bahwa Random Forest Classifier dan Gradient Boosting Classifier menunjukkan performa yang lebih baik dibandingkan dengan Logistic Regression. Kedua model ini berhasil mencapai accuracy score yang lebih tinggi menunjukkan bahwa kemampuannya dalam mengklasifikasikan data dengan lebih akurat. Hal ini menunjukkan bahwa model-model kompleks yang dibangun secara bertahap seperti Random Forest dan Gradient Boosting lebih efektif dalam menangani dataset yang kompleks dan non-linear dibandingkan dengan model linier seperti Logistic Regression.

Dan dari hasil accuracy score, dapat disimpulkan bahwa Gradient Boosting Classifier merupakan model yang terbaik dari ketiganya, karena GBC menggabungkan banyak model. Hal ini membuat GBC mampu mengurangi bias yang mungkin terjadi. Selain itu, GBC juga melakukan ensemble learning, dimana akan membantu mengurangi varians, sehingga model menjadi lebih stabil dan kurang rentan terhadap overfitting.
